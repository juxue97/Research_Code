```python
print("Current folder path:", cur_path)
dataset = []
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
for i,user in enumerate(os.listdir(cur_path)):
    image_folder = os.path.join(cur_path,user)
    print(i)
    #print(image_folder)
    os.chdir(image_folder)
    for face_images in os.listdir():
        face_imag = os.path.join(image_folder,face_images)

        
        image = cv2.imread(face_imag,1)
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 
                                              scaleFactor=1.3, 
                                              minNeighbors=5)
        
        for (x, y, w, h) in faces:
            roi = image[y:y+h, x:x+w]  # crop the face region

            # Display the cropped ROI for visualization
            #cv2.imshow('ROI', roi)
            #cv2.waitKey(0)
            # Close all windows
            #cv2.destroyAllWindows()
        image = cv2.resize(roi,(128,128))
        dataset.append([image,i])
```

```python
x=[]
y=[]
for features,label in dataset:
    x.append(features)
    y.append(label)
#.reshape(-1,128,128,3)
X = np.array(x).reshape(-1,128,128,3)
y = np.array(y).reshape(-1,1)
```
